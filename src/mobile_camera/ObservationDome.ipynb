# Observation Dome Visualization

Visualize the latitude and longitude grid points on an `ObservationDome` using Matplotlib.

```python
import sys
sys.path.append("..")
from exp_run_config import Config
Config.PROJECTNAME = "BerryPicker"

import matplotlib.pyplot as plt
from observation_dome import ObservationDome
```

```python
dome = ObservationDome(x=0.0, y=0.0, z=0.0, r=1.0, no_long=10, no_lat=10)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

dome.visualize(
    ax,
    pose_target=(dome.x, dome.y, dome.z),
)

plt.tight_layout()
plt.show()
```

```python
dome = ObservationDome(x=0.0, y=0.0, z=0.0, r=1.0, no_long=10, no_lat=10)

middle_subset = [
    (lat_index, long_index)
    for lat_index in range(3, 9)
    for long_index in range(3, 9)
]

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

dome.visualize(
    ax,
    lat_long_points=middle_subset,
    pose_target=(dome.x, dome.y, dome.z),
    title='Observation Dome Subset and End-Effector Poses',
)

plt.tight_layout()
plt.show()
```
